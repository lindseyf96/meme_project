```python
pop_minus_dc = pd.read_csv("pop_meme_per_state.csv")
pop_minus_dc
```

```python
# deleted d.c. since I had to get rid of it in the other df before

pop_minus_dc = pop_minus_dc[pop_minus_dc.Region != 'District of Columbia']
pop_minus_dc
```

```python
# merged both dfs 

final_map_df = pop_minus_dc.merge(df, left_on='Region', right_on='properties.abbrev')
final_map_df
```

```python
# new df in order to put the memes into lists

final_map_df1 = pop_minus_dc.merge(df, left_on='Region', right_on='properties.abbrev')
```

```python
# deleted a bunch of random useless columns

final_map_df1 = final_map_df1.drop(list(final_map_df1)[15:25], axis=1)
```

```python
# what is going to show up on the map

final_map_df1['properties.headline'] = final_map_df1['title']
final_map_df1['properties.article'] = '<b><h2>' + final_map_df1['title'].map(str) + '</h2></b><p>' + final_map_df1['about'].map(str) + '</p><p>' + 'State: ' + final_map_df1['Region'].map(str) + '</p><p>' + 'Origin: ' + final_map_df1['origin'].map(str) + '</p><p>' + 'Year Created: ' + final_map_df1['year'].map(str) + '</p><img src="' + final_map_df1['image'].map(str) + '" width="200">'   

final_map_df1
```

```python
# made new df with the top popularity per state and the frequency of memes

pop = final_map_df1.groupby(['Region', 'popularity']).size().reset_index(name='Freq')
pop
```

```python
# chose the colors for the map here

def make_color(num):
    if num == 1:
        return '#87FF99'
    elif num == 2:
        return '#64CBE8'
    else:
        return '#997AFF'


pop['properties.color'] = pop['Freq'].apply(lambda x: make_color(x))
```

```python
# combined the colors with the new df

pop
```

```python
# grouped by region and put everything going on the map in one column

join = pd.DataFrame(final_map_df1.groupby('Region')['properties.article'].apply(list).apply(lambda x: "<div id='Region'><p>%s</p></table>" % '</p><p>'.join(x)).reset_index(name='properties.article'))
join
```

```python
# joined above df with this one, where I grouped by meme for the headline function on the map

join1 = pd.DataFrame(final_map_df1.groupby('Region')['properties.headline'].apply(list).apply(lambda x: "<div id='Region'><p>%s</p></table>" % '</p><p>'.join(x)).reset_index(name='properties.headline'))
join1
```

```python
# almost there! product of last two df's:

pre_final = join.merge(join1, left_on='Region', right_on='Region')
pre_final
```

```python
# merged the newly created above df with the pop df from before that listed the top popularity and frequency

pre_final1 = pop.merge(pre_final, left_on='Region', right_on='Region')
pre_final1
```

```python
# merged the above df with the shape file

final = pre_final1.merge(df, left_on='Region', right_on='properties.abbrev')
final
```

```python
# deleted the geometry column since it was all null, and here's the FINAL df:

final = final.drop('geometry', axis=1)
final
```

```python
ok_json = json.loads(final.to_json(orient='records'))
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
geo_format = process_to_geojson(ok_json)
```
