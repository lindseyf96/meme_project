```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

import urllib
from IPython.display import display, Image
from IPython.core.display import HTML 

def display_image(url):
    response = urllib.request.urlopen(url).read()
    img = Image(response, width=200)
    display(img)
```

```python
rows = []

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

meme_list = ['doge', 'moth-lamp', 'distracted-boyfriend', 'i-crave-that-mineral', 'evil-kermit', 'expanding-brain', 'surprised-pikachu', 'mocking-spongebob', 'y-tho', 'spongegar-primitive-sponge-caveman-spongebob', 'drakeposting', 'grumpy-cat', 'arthurs-fist', 'salt-bae', 'spider-man-pointing-at-spider-man', 'is-this-a-pigeon', 'blinking-white-guy', 'confused-mr-krabs-when-you-wake-up-from-a-nap', 'dat-boi', 'why-the-fuck-you-lyin']


for meme in meme_list:
    url = f"https://knowyourmeme.com/memes/{meme}"
    response = requests.get(url, headers=headers)
    doc = BeautifulSoup(response.text)



    stuff = doc.find_all('div', class_="x_brb6 c")


    for s in stuff:

        row = {}

        title = doc.find_all('h1')
        row['title'] = title[0].text.strip()
        #print('---')
        year = doc.find_all('a', {'href': re.compile(r'/search[?]query[=]year(.*)')})
        row['year'] = year[0].text
        #print('---')
        origin = doc.find_all('dd', class_= "entry_origin_link")
        row['origin'] = origin[0].text.strip()
        #print('---')
        about = doc.find('h2', id='about').findNext('p')
        row['about'] = about.text.strip()
        #print('---')
        origin_summary = doc.find('h2', id='origin').findNext('p')
        row['origin summary'] = origin_summary.text.strip()
        #print('---')
        #image = doc.find_all('img', class_= 'wide')
        image = doc.find_all('img', {'data-src': re.compile(r'https://i.kym-cdn.com/entries/(.*)')})
        #image_link = image[0]['data-src']
        row['image'] = image[0]['data-src']
        #row['image'] = display_image(f"{image_link}")

        rows.append(row)
#print(rows)

df = pd.DataFrame(rows)

df
```
